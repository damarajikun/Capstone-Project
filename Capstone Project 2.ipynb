Dari hasil screening, diketahui bahwa data data untuk kolom Marital_Status tidak standard. Selain itu, data education jg kita simplifikasi untuk dapat memudahkan dalam pembacaan data

```python
# standarisasi data marital status
data['Marital_Status']=data['Marital_Status'].replace({'Divorced':'Alone','Single':'Alone','Married':'In couple','Together':'In couple','Absurd':'Alone','Widow':'Alone','YOLO':'Alone'})

# standatisasi data education
data['Education']=data['Education'].replace({'Basic':'Undergraduate','2n Cycle':'Undergraduate','Graduation':'Postgraduate','Master':'Postgraduate','PhD':'Postgraduate'})
data.head()
```

Pada grafik, nilai rata-rata transaksi konsumen postgraduate lebih tinggi dibanding konsumen undergraduate. Kita coba buktikan menggunakan metode statistika apakah nilai rata-rata transaksi konsumen postgraduste tersebut nilainya signifikan dibanding konsumen undergraduate.

```python
# Analisa menggunakan t-test untuk membuktikan rata-rata transaksi postgraduate dan undergraduate
# Ho = post <= under
# H1 = post > under

from scipy.stats import ttest_ind

store_postgrad= data['Store'].where(data['Education']=='Postgraduate').dropna()
store_undergrad= data['Store'].where(data['Education']=='Undergraduate').dropna()

stat, pval = ttest_ind(store_postgrad,store_undergrad, alternative='greater') 

if pval<=0.05:
    print('Tolak Ho')
else:
    print('Gagal menolak Ho')
```

Dari hasil uji statistika, didapatkan kesimpulan tolak H0 yang artinya tidak ada cukup bukti bahwa rata-rata transaksi konsumen undergraduate lebih besar atau sama dengan rata-rata transaksi konsumen postgraduate.

```python
# Rata-rata jumlah transaksi pada konsumen yang memiliki anak dan tidak
store_haskid= data['Store'].where(data['Has_kid']=='Has kid').dropna()
store_nokid= data['Store'].where(data['Has_kid']=='No kid').dropna()

print(f'Rata-rata jumlah transaksi konsumen yang memiliki anak = {store_haskid.mean()}')
print(f'Rata-rata jumlah transaksi konsumen yang memiliki anak = {store_nokid.mean()}')
```

Dari hasil perhitungan diatas, diketahui bahwa rata-rata jumlah transaksi untuk konsumen yang tidak memiliki anak lebih besar daripada konsumen yang memiliki anak. Kita coba buktikan secara statistik apakah nilai nya signifikan. 

```python
#Ho = haskid >= nokid
#H1 = haskid < nokid 
from scipy.stats import ttest_ind

stat, pval = ttest_ind(store_haskid,store_nokid, alternative='less') 

if pval<=0.05:
    print('Tolak Ho')
else:
    print('Gagal menolak Ho')
```

Dari hasil analisa menggunakan t-test, hasilnya menunjukkan bahwa ada perbedaan signifikan dimana rata-rata transaksi konsumen yang tidak memiliki anak leibh tinggi dari konsumen yang  memiliki anak.
